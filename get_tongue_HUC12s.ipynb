```python
# list of huc12's in the Tongue River Basin
hucs = pd.read_csv('tongue_r_huc12.csv', header=None)
hucs = hucs.iloc[:,0].to_list()
```

```python
try:
    os.makedirs("trhuc12_tiffs")
except FileExistsError:
    # directory already exists
    pass

for i in hucs:
    geom = pynhd.WaterData('wbd12').byid('huc12', str(i)).geometry[0]
    dem = py3dep.get_dem(geom, 30)
    dem.rio.to_raster(r'trhuc12_tiffs\{0}_30m.tif'.format(str(i)))
```
